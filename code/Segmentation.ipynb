# 3.2.1 Segmentation based on the short-time energy

The audio file is divided into overlapping frames.
In this work: frame size is 128 samples corresponding 3ms and adjacent frames overlap 50%.

```julia
using WAV, Plots

#x, fs = wavread("audio/snare_hit.wav");
x, fs = wavread("../dataset/RMBL-Robin/data/A-01june09-0702-robin.WAV")
x = x[:, 1];
#plotly();
#plot(x, linewidth=2, title= "Audio wave")
```

Energy envelope E(m) of the signal x(n) in decibel scale:
\begin{equation}
E(m) = \sum\limits_{i=1}^{N} 20log_{10}|x_i[n]|^2
\end{equation}
where $x_i[n]$ is i-th frame and $N$ is the total number of frames of the signal $x(n)$.

```julia
# Iterating through frames
# frame i, size N, hopsize H of signal x
#=
function frame_i(x, i, N, H)
    size_x = size(x)[1];
    if i*N <= size_x
        return x[1 + (i - 1)*N : i*N];
    else
        return x[1 + (i - 1)*N : size_x];
    end
end


function all_frames(x, N, H)
    x_size = size(x)[1];
    numberOfFrames = ceil(Int, x_size/N);
    for i =1:numberOfFrames
        print(frame_i(x, i, N, H))
    end
end

function energyEnvelope(x, N, H)
    x_size = size(x)[1];
    numberOfFrames = ceil(Int, x_size/N);
    E = zeros(numberOfFrames)
    for i = 1:numberOfFrames
        frame_ = frame_i(x, i, N, H); 
        for j = 1:N
            #println(N);
            #println(size(frame_))
            if i < N
                E[i] += 20*log10(frame_[j]*frame_[j]);
            else
                ub = size(frame_)[1];
                if j <= ub
                    E[i] += 20*log10(frame_[j]*frame_[j]);
                end
            end
        end
    end
    return E;
end
=#
```
